```python
import pandas as pd
```

```python
totals = pd.read_csv('totals.csv').set_index(keys=['name'])
```

```python
counts = pd.read_csv('counts.csv').set_index(keys=['name'])
```

```python
sum_rows_totals = totals.sum(axis = 1)
```

```python
min_city = sum_rows_totals.idxmin()
print("City with lowest total precipitation:")
print(min_city)
```

```python
sum_columns_totals = totals.sum(axis = 0)
sum_columns_counts = counts.sum(axis = 0)
```

```python
monthly_avg = sum_columns_totals / sum_columns_counts
print("Average precipitation in each month:")
print(monthly_avg)
```

```python
city_avg = sum_rows_totals / counts.sum(axis = 1)
```

```python
print("Average precipitation in each city:")
print(city_avg)
```
